```python
# Load the MovieLens 100K dataset
# Assuming the dataset is loaded into a DataFrame named 'df'
# You may need to adjust the file paths based on your directory structure
df = pd.read_csv('/kaggle/working/PMLDL_movie_recommender_system/data/interim/preprocessed_data.csv')
df
```

# Start training

```python
# Install Surprise library
!pip install scikit-surprise
```

```python
# Import necessary libraries
from surprise import Dataset, Reader, SVD, dump
from surprise.model_selection import cross_validate, train_test_split
from surprise import accuracy
```

```python
reader = Reader(rating_scale=(1, 5))
dataset = Dataset.load_from_df(df[['user_id', 'item_id', 'rating']], reader)

trainset, testset = train_test_split(dataset, test_size=0.2)
```

```python
# Choose the SVD algorithm (you can experiment with other algorithms)
model = SVD()

# Train the model on the training set
model.fit(trainset)
```

```python
# Make predictions on the test set
predictions = model.test(testset)

# Evaluate the model
accuracy.rmse(predictions)
```

```python
# Save the trained model
model_filename = 'model.pkl'
dump.dump(model_filename, algo=model)
```
